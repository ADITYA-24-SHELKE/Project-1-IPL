```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

```

```python
#ipl data loading
ipl=pd.read_csv('matchdata.csv')
```

```python
#overview of the matches data
ipl.head()
```

```python
#how many row and columns are there
ipl.shape
```

```python
#ipl madhle kon kon man of the match kiti vela jhalet
ipl['player_of_match'].value_counts()
```

```python
#top 11 pahije fkt
ipl['player_of_match'].value_counts()[0:11]
```

```python
list(ipl['player_of_match'].value_counts()[0:5].keys())
```

```python
plt.figure(figsize=(8,5))
plt.bar(ipl['player_of_match'].value_counts()[0:5].keys(),list(ipl['player_of_match'].value_counts()[0:5]),color='r')
plt.show()
```

```python
#ipl cha all match result
ipl['result'].value_counts()
```

```python
ipl['toss_winner'].value_counts()
```

```python
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
plt.figure(figsize=(5,5))
plt.hist(batting_first['win_by_runs'])
plt.title("distribution of runs")
plt.xlabel("runs")
plt.show()
```

```python
batting_first['winner'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color=["blue","yellow","orange"])
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f&&')
plt.show()
```
